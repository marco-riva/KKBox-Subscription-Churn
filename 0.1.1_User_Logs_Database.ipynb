```python
import numpy as np
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

```python
DIR = '.'
FILE  = '/user_logs.csv'
file = '{}{}'.format(DIR, FILE) 
```

```python
# read head of csv file
df_2_rows = pd.read_csv(file, nrows=2)
df_2_rows.columns
```

```python
# connector to database db
logs_db = create_engine('sqlite:///logs_db.db')
```

```python
# build db by chunking
chunk_size = 1000000
i = 0
j = 0

for df in pd.read_csv(file, chunksize=chunk_size, iterator=True):
    df = df.rename(columns = {c: c.replace(' ', '') for c in df.columns})
    df.index += j
    
    df.to_sql('data_use', logs_db, if_exists = 'append')
    j = df.index[-1] + 1
    
    print('| index: {}'.format(j))
```
